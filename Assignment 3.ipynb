Q8. Using NumPy, create a function to normalize the BloodPressure values (subtract the mean and divide by the standard deviation).

```python
def normalize(columnname):
    column_name = columnname.to_numpy()
    mean = np.mean(column_name)
    std = np.std(column_name)
    normalized = (column_name - mean)/std
    return normalized
```

```python
normalize(df['Glucose'])
```

### Matplotlib Questions
Q1. Plot a histogram of Age.

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.hist(df['Age'], bins=20, edgecolor='white')
ax.set_xlabel('Age')
ax.set_title("Distribution of Age")
plt.show()
```

Q2. Create a scatter plot showing Glucose versus Insulin levels.

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(df['Glucose'], df['Insulin'], alpha=0.7)
ax.set_title("Scatter Plot of Glucose vs Insulin Levels")
ax.set_xlabel("Glucose")
ax.set_ylabel("Insulin")
plt.show()
```

Q3. Draw a bar chart comparing the mean BMI across Outcome.

```python
d = df.groupby(['Outcome'])['Age'].mean()
fig, ax = plt.subplots(figsize=(6,4))
p = ax.bar(d.index,d.values,width = .9,color='skyblue', edgecolor="black")
ax.set_title('Mean BMI Across Outcome')
ax.set_xlabel('Outcome')
ax.set_ylabel('Mean BMI')
ax.set_xticks((0,1))
ax.bar_label(p)

plt.show()
```

Q5. Display a box plot of BMI grouped by Outcome.

Q6. Create a pie chart showing the proportion of diabetes outcomes (0's and 1's).

```python
x = df['Outcome'].value_counts()
label = df['Outcome'].value_counts().index

fig, ax = plt.subplots(figsize=(8,4))

ax.pie(x, labels=label,autopct='%1.1f%%', startangle=140)
ax.set_title('Proportion of Diabetes Outcomes')
ax.axis('equal') 
plt.show()
```

Q7. Plot a stacked bar chart of Pregnancies categorized by Outcome.

```python
pregnancies = df.groupby(['Pregnancies', 'Outcome']).size().unstack()

pregnancies.plot(kind='bar', stacked=True, figsize=(8, 4))
plt.title("Stacked bar chart of 'Pregnancies' categorized by 'Outcome'")
plt.xlabel('Pregnancies')
plt.ylabel('Outcome')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

bottom = None
for col in pregnancies.columns:
    ax.bar(x=pregnancies.index, height=pregnancies[col], bottom=bottom, label=col)
    if bottom is None:
        bottom = pregnancies[col].values
    else:
        bottom += pregnancies[col].values

ax.set_title('Stacked Bar Chart of Pregnancies Categorized by Outcome')
ax.set_xlabel('Pregnancies')
ax.set_ylabel('Count')
ax.set_xticks(pregnancies.index)

plt.legend(title='Outcome', loc='upper right')
plt.show()
```
